## Dropping some columns that are less relevant to the purpose of the analysis

```python
#Some variables have alot of missing values(>70%) and are not useful for this analysis. E.g. 'participant_name', 'participant_relationship.

#Some varibales don't have any missing values but are still not nedeed in this analysis. E.g. 'incident_url', incident_url_field_missing'

#variables like 'state_house_district', 'state_senate_district', 'congressional_district'  are too complicated to correct accurately.

#some variables have missing values but are useful in this analysis. E.g. 'address', 'gun_type', 'gun_stolen'

#i will start by dropping the variables that are not useful for the purpose of this analysis.
```

```python
cols_to_drop = ['incident_url', 'incident_url_fields_missing', 'sources','participant_name', 'participant_age',
                'participant_relationship', 'location_description', 'state_house_district', 'state_senate_district', 'congressional_district']
df_new = df.drop(columns=cols_to_drop)
```

```python
#check the new dataframe
df_new.shape
```

## Handle missing values 

```python
#check missing values in the new dataframe
df_new.isnull().sum()
```

```python
#Fill missing values in categorical and texts columns with "Unknown"
cat_col = ['address', 'source_url', 'gun_stolen', 'gun_type','incident_characteristics', 'notes', 'participant_age_group',
            'participant_gender', 'participant_status', 'participant_type']
for col in cat_col:
    df_new[col] = df_new[col].fillna('Unknown')

#fill missing values in numrical columns with the median   
num_col= ['n_guns_involved']
for col in num_col:
    df_new[col] = df_new[col].fillna(df_new[col].median())
```

```python
df_new.isnull().sum()
```

## Handle Missing Coordinates

```python
#Impute median latitude/longitude within each state
if 'longitude' in df_new.columns and 'latitude' in df_new.columns:
  df_new['longitude'] = df_new.groupby('state')['longitude'].transform(lambda x: x.fillna(x.median()))
df_new['latitude']  = df_new.groupby('state')['latitude'].transform(lambda x: x.fillna(x.median()))
```

```python
#check the result 
df_new.isnull().sum()
```

## Handle Datatype

```python
#convert 'n_guns_involved' column to integers because all the missing values has been fixed
df_new['n_guns_involved'] = df_new['n_guns_involved'].astype(int)

#convert date column to datetime format
df_new['date'] = pd.to_datetime(df_new['date'])
```

```python
#checking for mixed-data type in the dataframe
for col in df_new.columns.tolist():
  weird = (df_new[[col]].applymap(type) != df_new[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_new[weird]) > 0:
    print (col)
```
