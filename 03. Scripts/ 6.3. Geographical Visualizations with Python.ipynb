## **Create a choropleth Map**

**QUESTION 7**  
Create a choropleth map using a variable of your choice

**To answer questions based on severity (total incidence, total killed and total injured) per state**

```python
#rename the n_killed column to 'total killed' and n_injured to 'total_injured'
sub = sub.rename(columns = {'n_killed': 'total_killed','n_injured': 'total_injured', 'n_guns_involved': 'total_guns involved'})
```

```python
#check the result
sub.head()
```

```python
#Aggregate total killed and total injured by state (group by state)
agg = sub.groupby('state', as_index=False)[['total_killed', 'total_injured']].sum()
```

```python
agg.head()
```

```python
#total incidence per state 
agg["total_incidence"] = agg["total_killed"] + agg["total_injured"]
```

```python
#check result
agg.head()
```

```python
#create a choropleth map of total incidence per state

map = folium.Map(location = [40, -96], zoom_start =4)

folium.Choropleth(
    geo_data = country_geo, 
    data = agg,
    columns = ['state', 'total_incidence'],
    key_on = 'feature.properties.name', 
    fill_color = 'YlOrBr', fill_opacity=0.6, line_opacity=0.1,
    legend_name = "total incidence").add_to(map)
folium.LayerControl().add_to(map)

map
```

The choropleth map shows that gun violence incidents are heavily concentrated in a few states.  
California, Illinois, Texas, and Florida stand out with the highest total number of victims (killed + injured).  
Midwestern and Southern states also show moderately high incident levels, while many Western and Northeastern states display lower totals.  
Overall, the map highlights clear geographic disparities, with a small group of states accounting for a large share of national gun-related harm.

**To answer questions based on Gun metrics (total number of guns involved) per state**
