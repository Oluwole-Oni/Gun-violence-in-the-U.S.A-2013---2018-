## **CORRELLATION MATRIX HEATMAP**

**QUESTION 3**  
Create a correlation matrix heatmap (colored).

```python
#Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix
f,ax = plt.subplots(figsize=(10,10))
corr_heat = sns.heatmap(sub.corr(), annot = True, ax = ax) 
# The `annot` argument allows the plot to place the correlation coefficients onto the heatmap.

# Save the heatmap
plt.savefig(os.path.join(save_path, "correlation_heatmap.png"), dpi=300, bbox_inches='tight')
plt.show()
```

The heatmap displays the correlations among five numerical variables related to gun violence incidents: n_killed, n_injured, latitude, longitude, and n_guns_involved. The correlation values range from –1 to +1, where values closer to zero indicate no linear relationship.

**Relationship Between Fatalities and Injuries**  
The correlation between n_killed and n_injured is –0.13, suggesting a very weak negative relationship.This means the number of people injured in an incident does not reliably increase or decrease with the number killed.

**Guns Involved vs. Casualties**  
n_guns_involved shows extremely weak correlations with: n_killed (–0.022), n_injured (–0.033).
This means the number of firearms present does not strongly determine casualty counts, likely because:Some incidents involve multiple guns but few casualties. Others involve a single gun but cause significant harm.

**Location Variables Show No Meaningful Pattern**    
Latitude and longitude have near-zero correlations with: Number killed, Number injured, Number of guns involved
This indicates geographic coordinates do not predict severity of an incident on their own.

**Note**   
Gun violence is influenced by complex, non-linear factors, meaning that fatalities, injuries, guns used, and geographic coordinates do not demonstrate strong direct associations. This highlights the importance of incorporating contextual, demographic, and situational variables in further analysis.

In other words, incidents with a high number of deaths do not necessarily involve a high number of injuries, and vice versa.

Overall, high-fatality incidents tend to occur even when only a small number of guns are present, indicating that fatal outcomes are not driven by the quantity of guns involved but likely by other factors.

## **CATEGORICAL PLOT**

**QUESTION 6**  
Create a categorical plot and interpret the results

**Plot histogram distribution and create category**

```python
#plot an histogram for the distribution of n_kiiled to help craete the right categories for n_killed. 
plt.figure(figsize=(10,5))
sns.histplot(df_new['n_killed'], bins=20, kde=True)
plt.title("Distribution of Number Killed")
plt.xlabel("n_killed")
plt.ylabel("Frequency")
plt.show()
```

```python
#create killed category based on the distribuution.
df_new.loc[df_new['n_killed'] == 0, 'killed_category'] = 'No_death'
df_new.loc[(df_new['n_killed'] >= 1) & (df_new['n_killed'] <= 2), 'killed_category'] = 'Low_death'
df_new.loc[(df_new['n_killed'] >= 3) & (df_new['n_killed'] <= 5), 'killed_category'] = 'Medium_death'
df_new.loc[df_new['n_killed'] > 5, 'killed_category'] = 'High_death'
```

```python
df_new['killed_category'].value_counts(dropna = False)
```

```python
#plot an histogram for the distribution of n_injured to help create the right categories for n_injured. 
plt.figure(figsize=(10,5))
sns.histplot(df_new['n_injured'], bins=20, kde=True)
plt.title("Distribution of Number Injured")
plt.xlabel("n_injured")
plt.ylabel("Frequency")
plt.show()
```

```python
#create injured category based on the distribution.
df_new.loc[df_new['n_injured'] == 0, 'injured_category'] = 'No_injury'
df_new.loc[(df_new['n_injured'] >= 1) & (df_new['n_injured'] <= 3), 'injured_category'] = 'Low_injury'
df_new.loc[(df_new['n_injured'] >= 4) & (df_new['n_injured'] <= 10), 'injured_category'] = 'Medium_injury'
df_new.loc[df_new['n_injured'] > 10, 'injured_category'] = 'High_injury'
```

```python
df_new['injured_category'].value_counts(dropna = False)
```

```python
#clean the 'state' column: remove spaces and standardize capitalization
df_new['state'] = df_new['state'].str.strip().str.title()
```

```python
#check for missing or empty values
df_new = df_new[df_new['state'].notna() & (df_new['state'] != '')]
```

```python
#create top 20 states with the highest frequency
top_states_counts = df_new['state'].value_counts().head(20)
top_states = top_states_counts.index
```

```python
#check result
print("Top 20 states with counts:")
print(top_states_counts)
```

```python
#Filter the DataFrame for top states
df_top = df_new[df_new['state'].isin(top_states)]
```

```python
df_top.shape
```

```python
#create a categorical plot in seaborn using the kllled_category.
sns.set(style="ticks")

# Create the catplot 
g = sns.catplot(data=df_top,x="state",y="n_killed",hue="killed_category",kind="strip", jitter=True, height=6, aspect=2)

#customize labels and title
g.set_axis_labels("State", "Number Killed per Incident")
g.fig.suptitle("Killed per Incident Across Top 20 States", y=1.02)

plt.xticks(rotation=45)

# Save figure
plt.savefig(os.path.join(save_path, "state_killed_catplot.png"),
    dpi=300
)

plt.show()
```

This visualization examines the number of people killed per incident across the top states with the highest gun violence incident counts, categorized by death severity levels.   
The pattern across states is highly consistent, most incidents lead to no deaths or only a single fatality, as shown by the dense clusters of blue (No_death) and orange (Low_death) points.   
Medium_death cases (2–5 deaths) appear in every high-incident state but remain relatively limited in frequency.
Severe events involving six or more deaths are rare but significant. These (high_death) outliers represented by red points appear sporadically across states like California, Texas, Florida, and Missouri.   
Although uncommon, these spikes highlight isolated mass casualty events within the dataset.   
Importantly, no state shows a concentrated pattern of high_death incident, instead such events are scattered and exceptional.  
Overall, the plot shows that while gun violence incidents are widespread, fatal outcomes are not concentrated or frequent, and extreme events represent a small but impactful minority of cases.

```python
#create a categorical plot in seaborn using the injured_category.
sns.set(style="ticks")

# Create the catplot 
g = sns.catplot(data=df_top, x="state", y="n_injured", hue="injured_category",kind="strip", jitter=True, height=6, aspect=2)

#customize labels and title
g.set_axis_labels("State", "Number injured per Incident")
g.fig.suptitle("injured per Incident Across Top 20 States", y=1.02)

plt.xticks(rotation=45)

# Save figure
plt.savefig(os.path.join(save_path,"state_injured_catplot.png"), dpi=300)

plt.show()
```

The plot shows that most gun-violence incidents result in low to medium injury counts, with the majority falling between 0–8 injuries.    
While this pattern is consistent across all top 20 states, a few states particularly Florida, Texas, California, and New York display high-injury outliers, including rare incidents where many 20+ injuries were recorded.    
Overall, the distribution suggests that typical incidents are small-scale, but state trends are influenced by a small number of severe, high-injury events that significantly elevate the upper range of injuries.

**states with the highest incidents based on the number of people killed**

```python
#Group by state and sum the number killed
state_killed = df.groupby('state')['n_killed'].sum().reset_index()

# Sort states by number killed in descending order
state_killed_sorted = state_killed.sort_values(by='n_killed', ascending=False)

# Display top states with highest killed incidents
print(state_killed_sorted.head(20))
```

**states with the highest incidents based on the number of people injured**

```python
#Group by state and sum the number injured
state_injured = df.groupby('state')['n_injured'].sum().reset_index()

# Sort states by number killed in descending order
state_injured_sorted = state_injured.sort_values(by='n_injured', ascending=False)

# Display top states with highest injured incidents
print(state_injured_sorted.head(20))
```

**states with the highest incidents based on the number of people killed and injured**

```python
# Group by state and sum the number killed and injured
state_summary = df.groupby('state')[['n_killed', 'n_injured']].sum().reset_index()

# Sort states by killed and injured 
state_summary_sorted = state_summary.sort_values(by=['n_killed', 'n_injured'], ascending=False)

# Display top 10 states with highest killed and injured
print(state_summary_sorted.head(20))
```
